```python
import ollama
```

```python
response = ollama.chat(
    model="deepseek-r1:7b",
    messages=[
        {"role": "user", "content": "DeepSeek是什么？"},
    ],
    stream=True,
)
```

```python
print(response)
```

```python
print(response["message"]["content"])
```

```python
for chunk in response:
    print(chunk["message"]["content"], end="", flush=True)
```
